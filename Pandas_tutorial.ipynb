```python
import json

json_data = '''
[
    {"Company": "A", "ESG_Score": 75, "Metrics": {"Emission": 50, "Revenue": 200}},
    {"Company": "B", "ESG_Score": 60, "Metrics": {"Emission": 70, "Revenue": 300}},
    {"Company": "C", "ESG_Score": 90, "Metrics": {"Emission": 60, "Revenue": 250}}
]
'''
```

```python
# Convert JSON to DataFrame
data = json.loads(json_data)
df = pd.json_normalize(data)

```

```python
df
```

```python
def ecg_categories(x):
    if x > 80:
        return 'High'
    elif x >= 60 and x <= 80:
        return 'Medium'
    elif x < 60:
        return 'Low'
```

```python
df['ESG_Categories'] = df['ESG_Score'].apply(ecg_categories)
```

```python
df
```

```python
df['ESG_Categories1'] = df['ESG_Score'].apply(lambda x : 'High' if x > 80 else ('Medium' if x >= 60 else 'Low'))
```
