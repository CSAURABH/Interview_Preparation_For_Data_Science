### Label Encoding

```python
df = pd.read_csv(r'C:\Users\HP\Desktop\Interview_Preparation_For_Data_Science\Practice\train.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
df['Cabin'] = df['Cabin'].fillna(df['Cabin'].mode()[0])
```

```python
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df.info()
```

```python
columns = df.select_dtypes(include='object')
```

```python
columns.drop('Name', inplace=True, axis = 1)
```

```python
columns
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
for column in columns:
    lencoder = LabelEncoder()
    df[column] = lencoder.fit_transform(df[column])
```

```python
df.head()
```

### Ordinal Encoding

```python
df = pd.read_csv(r'C:\Users\HP\Desktop\Interview_Preparation_For_Data_Science\Practice\customer.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.review.value_counts()
```

```python
df.education.value_counts()
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
odencodingReview = OrdinalEncoder(categories=[['Poor','Good','Average']])
df['review'] = odencodingReview.fit_transform(df[['review']])
```

```python
odencodingEducation = OrdinalEncoder(categories=[['PG','School','UG']])
df['education'] = odencodingEducation.fit_transform(df[['education']])
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
columns = df.select_dtypes(include='object')
```

```python
for col in columns:
    lencoder = LabelEncoder()
    df[col] = lencoder.fit_transform(df[col])
```
